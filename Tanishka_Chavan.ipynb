```python
import pandas as pd
df = pd.read_csv("data/ final_food_delivery_dataset.csv")
df.head()
```

```python
gold_members = df[df['membership'] == 'Gold']
city_revenue = gold_members.groupby('city')['total_amount'].sum()
top_cities = city_revenue.sort_values(ascending=False)
print(top_cities.head(1))
```

```python
cuisine_highest_average_order_value = df.groupby('cuisine')['total_amount'].mean()
top_cuisines = cuisine_highest_average_order_value.sort_values(ascending=False)
print(top_cuisines.head(1))
```

```python
cuisine_avg_value = df.groupby('cuisine')['total_amount'].mean()
top_cuisines = cuisine_avg_value.sort_values(ascending=False)
print(top_cuisines.head(1))
```

```python
users_total_spending = df.groupby('user_id')['total_amount'].sum()
total_top_users = users_total_spending[users_total_spending > 1000]
print(len(total_top_users))
```

```python
bins = [2.9, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0 – 3.5', '3.6 – 4.0', '4.1 – 4.5', '4.6 – 5.0']
df['rating_range'] = pd.cut(df['rating'], bins=bins, labels=labels)
revenue_by_rating = df.groupby('rating_range')['total_amount'].count()
print(df.groupby('rating')['total_amount'].count())
print(revenue_by_rating.sort_values(ascending=False))
```

```python
gold_city_avg = gold_members.groupby('city')['total_amount'].mean()
top_gold_city = gold_city_avg.sort_values(ascending=False)
print(top_gold_city.head(1))
```

```python
cuisine_stats = df.groupby('cuisine').agg({
    'restaurant_id': 'nunique', 
    'total_amount': 'sum'
})
cuisine_stats.columns = ['distinct_restaurants', 'total_revenue']
result = cuisine_stats.sort_values(by='distinct_restaurants', ascending=True)
print(result)
```
